First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
URL = 'datasets/spacex_launch_geo.csv'
spacex_df=pd.read_csv(URL)
spacex_df.head()
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
## Task 1: Mark all launch sites on a map
# launch_sites_df contains all the latitudes and longitudes for all the launch sites

# limit the map to show america
world_map = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Create a marker for each launch site in the dataframe
for index, row in launch_sites_df.iterrows():
    folium.Marker([row['Lat'], row['Long']], popup=row['Launch Site']).add_to(world_map)

# Display the map
world_map
```

```python
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# a list for circles and markers
circles = []
markers = []

# loop through each launch site
for index, row in launch_sites_df.iterrows():
    # create a circle for each launch site
    circle = folium.Circle([row['Lat'], row['Long']], radius=1000, color='#d35400', fill=True).add_child(folium.Popup(row['Launch Site']))
    # create a marker for each launch site
    marker = folium.map.Marker(
        [row['Lat'], row['Long']],
        # create an icon as a text label
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % row['Launch Site'],
            )
        )
    # add the circle and marker to the list
    circles.append(circle)
    markers.append(marker)

# add the circles and markers to the map
for circle in circles:
    site_map.add_child(circle)

for marker in markers:
    site_map.add_child(marker)  

site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# Create a new column called `marker_color` and give it default value `gray`
spacex_df['marker_color'] = 'gray'
spacex_df.head()

# Update the marker_color column and assign a color based on its class
spacex_df.loc[spacex_df['class'] == 1, 'marker_color'] = 'green'
spacex_df.loc[spacex_df['class'] == 0, 'marker_color'] = 'red'

# create a list for markers
markers = []

# print the first row of the dataframe
# print(spacex_df.iloc[0])

# loop through the launches in each launch site
for launch in spacex_df.values:
    # create a marker for each launch
    marker = folium.map.Marker(
        [launch[1], launch[2]],
        # add a popup label showing the Launch Site for each launch, according to the color of launch[4]
        popup=launch[0],
        icon=folium.Icon(color=launch[4])
    )
        
    # add the marker to the list
    markers.append(marker)

# create a new map to show the launch sites
launch_map = folium.Map(location=nasa_coordinate, zoom_start=5)

for marker in markers:
    launch_map.add_child(marker)

launch_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object

```python
# create a MarkderCluster object
marker_cluster = MarkerCluster()

# Add marker_cluster to a new site_map
cluster_map = folium.Map(location=nasa_coordinate, zoom_start=5)   
cluster_map.add_child(marker_cluster)

# for each row in spacex_df data frame
# create a Marker object with its coordinate
# and customize the Marker's icon property to indicate if this launch was successed or failed, 
# e.g., icon=folium.Icon(color='white', icon_color=row['marker_color']
for index, record in spacex_df.iterrows():
    # TODO: Create and add a Marker cluster to the site map
    marker = folium.Marker(
        [record['Lat'], record['Long']],
        popup=record['Launch Site'],
        icon=folium.Icon(color='white', icon_color=record['marker_color'])
    )
    marker_cluster.add_child(marker)

for index, row in launch_sites_df.iterrows():
    # create a circle for each launch site
    circle = folium.Circle([row['Lat'], row['Long']], radius=1000, color='#d35400', fill=True).add_child(folium.Popup(row['Launch Site'])).add_to(cluster_map)

cluster_map
```

From the color-labeled markers in marker clusters, you should be able to easily identify which launch sites have relatively high success rates.


```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Import the geodesic module from the library
# %pip install geopy
from geopy.distance import geodesic

# Create a list to store the distances to the closest launch site
distances = []

# loop through each launch site
for index, row in launch_sites_df.iterrows():
    # create a list to store the distances to the closest launch site
    distances_to_launch_site = []
    # loop through each launch
    for launch in spacex_df.values:
        # calculate the distance between a launch site to its proximities
        distance = geodesic((row['Lat'], row['Long']), (launch[1], launch[2])).km
        # add the distance to the list
        distances_to_launch_site.append(distance)
    # add the closest distance to the list
    distances.append(min(distances_to_launch_site))

# print the list
print(distances)


```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
